# Desafio: Classificação de Subsegmentos Food Service

## Contexto

Uma empresa do setor de dados B2B precisa classificar estabelecimentos do ramo alimentício em **subsegmentos** a partir de dados cadastrais e features externas. Os subsegmentos-alvo são:

- **CASH AND CARRY** (atacarejos)
- **AUTO SERVICO** (supermercados/minimercados)
- **LANCHONETE**
- **CONFEITARIA**
- **PADARIA**

Você recebeu um conjunto de tabelas na pasta `outputs/` que contêm dados rotulados e diversas fontes de features. Seu objetivo é construir um pipeline de classificação que atinja a melhor performance possível.

## Dados Disponíveis

Na pasta `outputs/` você encontrará os seguintes arquivos:

| Arquivo | Descrição |
|---------|----------|
| `food.parquet` | Dataset rotulado com CNPJs, subsegmento e partição train/test |
| `places.parquet` | Dados do Google Places vinculados a CNPJs |
| `ifood.parquet` | Dados de restaurantes do iFood |
| `eb.parquet` | Dados cadastrais e financeiros de empresas brasileiras |
| `cnaes.parquet` | Tabela de referência: códigos CNAE e suas descrições |
| `divida.parquet` | Dados de dívida ativa agregada por CNPJ |

O dicionário detalhado de cada coluna está no arquivo `dicionario.md`.
